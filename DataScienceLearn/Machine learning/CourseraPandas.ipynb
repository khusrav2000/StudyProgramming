```python
def get_first_name(name):
    first_name = ""
    p=False
    x=False
    g = '(' if '(' in name else '.'
    for c in name:
        if p:
            if (c >= 'A' and c <= 'Z' ) | (c >= 'a' and c <= 'z'):
                x = True
            else:
                if x:
                    break
                x = False
        first_name += c if x else ''     
        if c == g:
            p=True
            
    print(name, first_name)
    return first_name
    
```

```python
first_names = []
for name in titanic.Name:
    first_names.append(get_first_name(name))
```

```python
titanic['first_name'] = first_names
```

```python
titanic[titanic.Sex == 'female'].groupby('first_name', as_index=False)\
    .agg({'Sex': 'count'}).sort_values("Sex")
```

```python
submissions_data_train = submissions_data_train.pivot_table(index = 'user_id', 
                                       columns='submission_status', 
                                       values='step_id', 
                                       aggfunc='count', 
                                       fill_value=0).reset_index()
```

```python
event_data_train = event_data_train.pivot_table(index = 'user_id', 
                                       columns='action', 
                                       values='step_id', 
                                       aggfunc='count', 
                                       fill_value=0).reset_index()
```

```python
event_data_train
```

```python
event_data_train.user_id.nunique()
```

```python
data = event_data_train.merge(submissions_data_train, how='outer')
```

```python
data = data.fillna(0)
```

```python
data['done'] = [1 if i >= 40.0 else 0 for i in data['correct']]
```

```python
data
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
X_train = data.drop(['done'], axis = 1)
y_train = data['done']
```

```python
clf_rf = RandomForestClassifier()
```

```python
parametrs = {'n_estimators': range(10, 50, 10),
             'max_depth': range(1, 12, 2),
             'min_samples_leaf': range(1, 7),
             'min_samples_split': range(2, 9, 2)}
```

```python
grid_search_cv = GridSearchCV(clf_rf, parametrs, cv=3, n_jobs=4)
```

```python
grid_search_cv.fit(X_train, y_train)
```

```python
best_clf = grid_search_cv.best_estimator_
```

```python
grid_search_cv.best_params_
```

```python
best_clf.score(X_train, y_train)
```

```python
submission_data_test = submission_data_test.pivot_table(index = 'user_id', 
                                       columns='submission_status', 
                                       values='step_id', 
                                       aggfunc='count', 
                                       fill_value=0).reset_index()
events_data_test = events_data_test.pivot_table(index = 'user_id', 
                                       columns='action', 
                                       values='step_id', 
                                       aggfunc='count', 
                                       fill_value=0).reset_index()
```

```python
data_test = events_data_test.merge(submission_data_test, how='outer')
```

```python
X_test = data_test.fillna(0)
```

```python
predictions = best_clf.predict_proba(X_test)
```

```python
predictions
```

```python
predict_df = pd.DataFrame({'user_id': X_test['user_id'], 'is_gone': pd.Series(predictions[:, 1])})
```
